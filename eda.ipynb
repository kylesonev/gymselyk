```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
```

```python
data = pd.read_csv("data/alunos.csv")
```

```python
data.head()
```

```python
print("Informações do dataset")
print(f"Número de Colunas: {data.shape[1]}")
print(f"Número de Amostras: {data.shape[0]}")
print(f"Média de Idade: {data["idade"].mean()}")
```

```python
contagem_sexo = data["sexo"].value_counts()
plt.figure(figsize=(16, 8))
sns.barplot(x=contagem_sexo.index, y=contagem_sexo.values, hue=contagem_sexo.index, palette="Set1")
plt.title("Distribuição por Sexo", fontsize=20)
plt.show()

```

```python
contagem_split = data["split"].value_counts()
plt.figure(figsize=(20, 8))
sns.barplot(x=contagem_split.index, y=contagem_split.values, hue=contagem_split.index, palette="Set1")
plt.title("Distribuição por Sexo", fontsize=20)
plt.xlabel("Splits", fontsize=15)
plt.xticks(fontsize=10, rotation=90)
plt.show()
```

```python
plt.figure(figsize=(16, 8))
sns.histplot(data=data, x=data["idade"], color="green")
plt.title("Distribuição por Idade")
plt.xlabel("Idade")
plt.ylabel("Número de pessoas")
plt.show()
```

```python
data["nivel"].value_counts()
```

```python
nivel = data["nivel"].value_counts()
nivel.index = ["Intermediário", "Iniciante", "Avançado"]
plt.figure(figsize=(16, 8))
sns.barplot(x=nivel.index, y=nivel.values, hue= nivel.index, palette="Set1")
plt.title("Contagem de Nível de Treino", fontsize=22)
plt.xlabel("Nível")
plt.ylabel("Número")
plt.show()
```
